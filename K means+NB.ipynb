```python
#import data and calculate standard score
import numpy as np
from keras.datasets import cifar10
(x_train, y_train), (x_test, y_test)=cifar10.load_data()
features=np.concatenate((x_train,x_test),axis=0)
labels=np.concatenate((y_train,y_test),axis=0).reshape(1,-1)[0].tolist()
features=features.astype('float')
features -= np.mean(features, axis=0)
features /= np.std(features, axis=0)
```

```python
features.shape
```

```python
#import image extraction library
from sklearn.cluster import MiniBatchKMeans
from sklearn.feature_extraction.image import extract_patches_2d
rng = np.random.RandomState(0)
kmeans = MiniBatchKMeans(n_clusters=1200, random_state=rng, verbose=True)
```

```python
#feature learning
buffer = []
for i in range(60000):
    img=features[i]
    data = extract_patches_2d(img, (7,7), max_patches=20,random_state=rng)
    data = np.reshape(data, (len(data), -1))
    buffer.append(data)
    if (i % 100 == 0 and i!=0):
        data = np.concatenate(buffer, axis=0)
        kmeans.partial_fit(data)
        buffer = []
    if i % 1000 == 0:
        print('Partial fit of %4i out of %i'
              % (i, 60000))
```

```python
#feature extraction
result = []
buffer=[]
for i in range(60000):
    img=features[i].astype('float')
    data = extract_patches_2d(img, (7,7))

    data = np.reshape(data, (len(data), -1))
    buffer.append(data)

    if ((i+1) % 100 == 0 and i!=0):
        data = np.concatenate(buffer, axis=0)
        predict_result_for_100=kmeans.predict(data).reshape(100,-1)
        for j in range(100):
          labels_676=predict_result_for_100[j]
          a01=np.zeros(1200)
          a02=np.zeros(1200)
          a03=np.zeros(1200)
          a04=np.zeros(1200)
          a1=np.bincount(labels_676[0:169])
          a01[:a1.shape[0]]=a1
          a2=np.bincount(labels_676[169:338])
          a02[:a2.shape[0]]=a2
          a3=np.bincount(labels_676[338:507])
          a03[:a3.shape[0]]=a3
          a4=np.bincount(labels_676[507:676])
          a04[:a4.shape[0]]=a4
          tem_result=np.concatenate((a01,a02,a03,a04),axis=0)
          result.append(tem_result)
               
        buffer = []
    if (i+1) % 100 == 0:
        print('Partial fit of %4i out of %i'
              % (i, 60000))
```

```python
new_result=np.asarray(result)
```

```python
#NB model
from sklearn.naive_bayes import MultinomialNB
def cal_accurancy(train_features,train_labels,test_features,test_labels):
    clf=MultinomialNB().fit(train_features,train_labels)
    result=clf.predict(test_features)
    n=0
    for i in range(6000):
        confusion[result[i]][test_labels[i]]+=1
        if(result[i]==test_labels[i]):
            n=n+1

    for i in range(10):
        for j in range(6000):
            if(result[j]==i and test_labels[j]==i):
              TP[i]+=1
            if(result[j]==i and test_labels[j]!=i):
              FP[i]+=1
            if(result[j]!=i and test_labels[j]==i):
              FN[i]+=1
            if(result[j]!=i and test_labels[j]!=i):
              TN[i]+=1
    return n/6000
```

```python
features_PCA=new_result
```

```python
#10-fold cross validation
TP=np.zeros(10)
FP=np.zeros(10)
TN=np.zeros(10)
FN=np.zeros(10)
Accuracy=np.zeros(10)
Precision=np.zeros(10)
Recall=np.zeros(10)
F1_Score=np.zeros(10)
confusion=np.zeros((10,10))
acc=[]
for i in range(10):
    train_features=np.concatenate((features_PCA[0:i*6000], features_PCA[(i+1)*6000:]), axis=0)    
    train_labels=labels[0:i*6000]+labels[(i+1)*6000:]
    test_features=features_PCA[i*6000:(i+1)*6000]
    test_labels=labels[i*6000:(i+1)*6000]
    acc.append(cal_accurancy(train_features,train_labels,test_features,test_labels))
    
print(sum(acc)/10)
for i in range(10):
    Accuracy[i]=(TP[i]+TN[i])/(TP[i]+TN[i]+FN[i]+FP[i])
    Precision[i] = TP[i]/(TP[i]+FP[i])
    Recall[i] = TP[i]/(TP[i]+FN[i])
    F1_Score[i] = 2*(Recall[i] * Precision[i]) / (Recall[i] + Precision[i])
```

```python
print(Accuracy)
print(Precision)
print(Recall)
print(F1_Score)
```
